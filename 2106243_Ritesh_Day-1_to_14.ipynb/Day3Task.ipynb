Q1) What are Data Structures? List some of data structures and their use in real world application?

Data structures are fundamental components of computer science that organize and store data in a specific way to enable efficient operations and access. They provide a way to manage and manipulate data effectively, optimizing tasks such as insertion, deletion, and retrieval.

Here are some commonly used data structures along with their real-world applications:



1.  **Arrays:** Arrays store a fixed-size sequence of elements of the same data type. They are used in various applications, such as storing a list of student's grades, representing images, and implementing matrices.

2.   **Linked Lists:** Linked lists consist of nodes that contain data and a reference to the next node. They are useful for dynamic data structures where elements can be easily inserted or removed. Linked lists are employed in applications like implementing stacks, queues, and managing memory in operating systems.

3.   **Stacks:** Stacks follow the Last-In-First-Out (LIFO) principle. They are used for tasks like parsing expressions, implementing undo-redo functionality, and managing function calls in programming languages.

4.   **Queues:** Queues follow the First-In-First-Out (FIFO) principle. They are suitable for scenarios such as job scheduling, handling requests in web servers, and buffering data in networking.

5.  **Trees:** Trees are hierarchical data structures consisting of nodes connected by edges. They have applications in file systems, decision-making processes, organization structures, and searching algorithms like binary search trees.

Q4) What is difference between ordered data type and unordered data type?

**Ordered data types**

*   A order data type maintains the order of elements as they are inserted.
*   Each element in the data structure has a specific position or index.
*   Examples of ordered data types in Python include lists and tuples.  

**Unordered data types**

*   An unordered data type does not maintain the order of elements.
*   The elements are stored in a way that allows for efficient lookup and retrieval, but their specific order is not guaranteed.
*   Examples of unordered data types in Python include sets and dictionaries.
